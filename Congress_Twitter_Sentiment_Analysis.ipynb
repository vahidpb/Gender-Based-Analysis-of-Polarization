```python
congress_tweets=pd.read_csv(r'C:\Users\vahidpb\Documents\GitHub\Gender-Based-Analysis-of-Polarization\congress_tweets.csv')
for i in userlist_new:
    #username=i
    try:
        # Converting whole text into lower case to maintain uniformity
        congress_tweets[i]=congress_tweets[i].str.lower()
        # Eliminating the punctuations
        congress_tweets[i]=congress_tweets[i].str.replace('[^\w\s]', '') 
        # Eliminating the numbers
        congress_tweets[i]=congress_tweets[i].str.replace('\d', '')
        # Eliminating the stop words
        congress_tweets[i]=congress_tweets[i].apply(lambda x: " ".join(x for x in str(x).split() if x not in stop_words))
        # Converting the text column as a single string for wordcloud
        reviews=str(congress_tweets[i])
        # Removing the URLS from the tweet string
        reviews_new=re.sub(r'http\S+', ' ', reviews)
        # Deleting everything which is not characters
        reviews_new = re.sub(r'[^a-z A-Z]', ' ',reviews_new)
        # Stripping extra spaces in the text
        reviews_new= re.sub(r' +', ' ', reviews_new)
        # Creating the custom stopwords
        customStopwords = stop_words
    except Exception as e:
        print(f'{i}')
```

# Applying Sentiment Intensity Analyzer

```python
# Creating an instance of the method
   
analysis_sa=SentimentIntensityAnalyzer()
for i in userlist_new:
    compoundlist=[]
    sentiment=[]
    for j in congress_tweets[i]:
        result=analysis_sa.polarity_scores(j)
        compoundlist.append(result['compound'])
        if result['compound']>0:
            temp='positive'
        if result['compound']<0:
            temp='negative'
        if result['compound']==0:
            temp='neutral'
        sentiment.append(temp)
    congress_tweets.loc[:,'compound_'+i]=compoundlist
    congress_tweets.loc[:,'sentiment_'+i]=sentiment
#print(congress_tweets.head(10))
congress_tweets.to_csv('new_congress_tweets.csv', index=False)
```
